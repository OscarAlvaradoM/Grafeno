# Graficas Raman

```julia
using LaTeXStrings
```

```julia
using Plots
```

```julia
using CSV
```

## Cobre

```julia
grafeno_Oscar_4 = CSV.read("grafeno_Oscar_4.CSV"; datarow=1)
Pd060319_1_1 = CSV.read("Pd060319_1_1.CSV"; datarow=1)

p1 = plot(grafeno_Oscar_4[:,1], 
    grafeno_Oscar_4[:,2], 
    ylabel = L"\textbf{Intensidad}", 
    xlabel = L"\textbf{Espectro\ de\ Raman} (cm^{-1})", 
    title = "Espectro Raman de Grafeno sobre Cobre", 
    leg = false, lc = :red,
    xticks = 1000:200:3500,
    yticks = 0:2:16)

#savefig(p1, "Espectro_Raman_Cobre.png")
```

## Paladio

```julia
Pd060319_1_1 = CSV.read("Pd060319_1_1.CSV"; datarow=1)

p2 = plot(Pd060319_1_1[:,1], 
    Pd060319_1_1[:,2], 
    ylabel = L"\textbf{Intensidad}", 
    xlabel = L"\textbf{Espectro\ de\ Raman} (cm^{-1})", 
    title = "Espectro Raman de Grafeno sobre Paladio", 
    leg = false, lc = :blue,
    xticks = 1000:200:3500,
    yticks = 0:30:300)

#savefig(p2, "Espectro_Raman_Paladio.png")
```

## Paladio y cobre

```julia
a = plot(p1, p2, layout = (2,1))

savefig(a, "Espectro_Raman.png")
```
