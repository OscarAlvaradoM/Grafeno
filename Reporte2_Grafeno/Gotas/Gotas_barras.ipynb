## Paladio

```julia
gotas_resultado = CSV.read("gotas_resultado.csv"; datarow = 2)

nam = [0,120,240,360,480,600,720,840,960,1080,1200,1320,1440]

a = [gotas_resultado[1:13,2] gotas_resultado[1:13,3]]
bar_paladio = groupedbar(nam, 
    a, 
    bar_position = :dodge, 
    bar_width = 60, 
    title = "Ángulo de corte Paladio", 
    xlabel = "Tiempo[s]", 
    ylabel = "Ángulo [°C]",
    ylims = [0,100],
    xticks = 0:120:1440,
    yticks = 0:10:100,
    lab = ["Con grafeno", "Sin grafeno"])

#savefig(bar_paladio, "Angulo_Corte_Paladio.png")
```

# Trabajo de adhesión

## Paladio

```julia
gotas_resultado = CSV.read("gotas_resultado.csv"; datarow = 2)

plot_paladio = plot(
    [gotas_resultado[32:44,1], gotas_resultado[32:44,1]], 
    [gotas_resultado[32:44,2], gotas_resultado[32:44,3]],
    title = "Trabajo de adhesión Paladio", 
    xlabel = "Tiempo[s]", 
    ylabel = L"\textbf{Trabajo} [J/m^{2}]",
    ylims = [0,0.16],
    xticks = 0:120:1440,
    yticks = 0:0.02:0.16,
    lab = ["Con grafeno", "Sin grafeno"],
    lc = [:blue :red],
    legend = :bottomright,
    marker=([:hex :d],6,0.8,stroke(1,:gray)),
    lw = 4)

#savefig(plot_paladio, "Trabajo_Adhesión_Paladio.png")
```

## Cobre

```julia
gotas_resultado = CSV.read("gotas_resultado.csv"; datarow = 2)

plot_cobre = plot(
    [gotas_resultado[32:44,6], gotas_resultado[32:44,6]], 
    [gotas_resultado[32:44,8], gotas_resultado[32:44,7]],
    title = "Trabajo de adhesión Cobre", 
    xlabel = "Tiempo[s]", 
    ylabel = L"\textbf{Trabajo} [J/m^{2}]",
    ylims = [0,0.16],
    xticks = 0:120:1440,
    yticks = 0:0.02:0.16,
    lab = ["Con grafeno", "Sin grafeno"],
    lc = [:yellow :purple],
    legend = :bottomright,
    marker=([:hex :d],6,0.8,stroke(1,:gray),[:yellow :purple]),
    lw = 4)

#savefig(plot_cobre, "Trabajo_Adhesion_Cobre.png")
```
