## Paladio

```julia
gotas_resultado = CSV.read("gotas_resultado.csv"; datarow = 2)

nam = [0,120,240,360,480,600,720,840,960,1080,1200,1320,1440]

a = [gotas_resultado[1:13,2] gotas_resultado[1:13,3]]
bar_paladio = groupedbar(nam, 
    a, 
    bar_position = :dodge, 
    bar_width = 110, 
    title = "Ángulo de corte Paladio", 
    xlabel = "Tiempo[s]", 
    ylabel = "Ángulo [°]",
    ylims = [0,100],
    xticks = 0:120:1440,
    yticks = 0:10:100,
    lab = ["Con grafeno", "Sin grafeno", font(color = RGB(255/255,241/255,98/255))],
    c = RGB(166/255,169/255,168/255),
    bg = RGB(29/255, 39/255, 74/255),
    titlefont = font("Dosis", 20, color = RGB(255/255,241/255,98/255)),
    tickfont = font(10, "Dosis", color = RGB(255/255,241/255,98/255)),
    guidefont = font(15, "Dosis",  color = RGB(255/255,241/255,98/255)))


savefig(bar_paladio, "/home/oscar/Dropbox/Grafeno/Congreso/Angulo_Corte_Paladio.png")
```

# Trabajo de adhesión

## Paladio

```julia
gotas_resultado = CSV.read("gotas_resultado.csv"; datarow = 2)

plot_paladio = plot(
    [gotas_resultado[32:44,1], gotas_resultado[32:44,1]], 
    [gotas_resultado[32:44,2], gotas_resultado[32:44,3]],
    title = "Trabajo de adhesión Paladio", 
    xlabel = "Tiempo[s]", 
    ylabel = "Trabajo [J/m²]",
    ylims = [0,0.16],
    xticks = 0:120:1440,
    yticks = 0:0.02:0.16,
    lab = ["Con grafeno", "Sin grafeno"],
    lc = [RGB(255/255,241/255,98/255) RGB(166/255,169/255,168/255)],
    legend = :bottomright,
    marker=([:hex :d],9,0.9,stroke(1,:black), [RGB(255/255,241/255,98/255) RGB(166/255,169/255,168/255)]),
    lw = 7,
    bg = RGB(29/255, 39/255, 74/255),
    titlefont = font("Dosis", 20, color = RGB(255/255,241/255,98/255)),
    tickfont = font(10, "Dosis", color = RGB(255/255,241/255,98/255)),
    guidefont = font(15, "Dosis", color = RGB(255/255,241/255,98/255)))

savefig(plot_paladio, "/home/oscar/Dropbox/Grafeno/Congreso/Trabajo_Adhesión_Paladio.png")
```
