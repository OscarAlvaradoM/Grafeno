```julia
Steren = [30, 30, 29, 29, 29, 29, 29, 29, 29, 29, 29, 26, 26, 25, 25, 24, 23, 22]
Termistor = [610, 603, 608, 607, 604, 607, 608, 604, 603, 607, 602, 567, 566, 562, 559, 558, 546, 535]

scatter(Termistor, Steren, leg = false)

Modelo(x,p) = p[1].*x.+p[2]
y = Steren
x = Termistor
p0 = [0.0, 0.0]

Ajuste = curve_fit(Modelo, x, y, p0)

Y = Ajuste.param

p1 = plot!(x,Modelo(x,Y),
    xlabel = "Mediciones de voltaje tomadas con el termistor",
    ylabel = "Termopar Steren",
    title = "Ajuste lineal Steren - Termistor")
annotate!([(562,29,text(" f(x) = $(Y[1]) * x + $(Y[2]) ",10,:black,:center))])

#print(Y[1],"\n",Y[2])
```
